```python
# project on Numpy
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open('s.jpg')
img_arr = np.array(img)
plt.imshow(img_arr)
```

```python
img_arr[0:175,0:325, 1]=0

img_arr[0:175,325:, 0]=0
img_arr[0:175,325:, 1]=0

img_arr[175:,0:325, 0]=0
img_arr[175:,0:325, 2]=0

img_arr[175:,325:, 1]=0
img_arr[175:,325:, 2]=0

plt.imshow(img_arr)
```

## Project Summary: Image Manipulation with NumPy and Matplotlib
Objective: The goal of this project was to perform image manipulation using NumPy arrays and visualize the results with Matplotlib.

### Description:

#### Image Loading:
The project begins by loading an image (s.jpg) using the Python Imaging Library (PIL) and converts it into a NumPy array. This array represents the image data in terms of pixel values.

#### Initial Visualization: 
The original image array is displayed using Matplotlib's imshow function to provide a visual reference before any modifications are applied.

#### Image Manipulation:

The image is divided into four quadrants.
Top-left Quadrant: The green channel (second channel in RGB) is set to zero, effectively removing the green color from this section.
Top-right Quadrant: Both the red and green channels are set to zero, resulting in a section with only blue hues.
Bottom-left Quadrant: Both the red and blue channels are set to zero, making this section predominantly green.
Bottom-right Quadrant: The green and blue channels are set to zero, leaving only the red hues in this section.
Final Visualization: The manipulated image array is displayed using Matplotlib to show the effects of the color modifications across the four quadrants.

Techniques and Tools:

#### NumPy:
Used for array manipulation, which allows direct modification of pixel values in the image.
#### Matplotlib:
Used for visualizing the image before and after manipulation.
#### PIL:
Utilized for reading the image file and converting it into a NumPy array.
